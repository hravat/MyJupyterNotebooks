# The Central Limit Theorem<br>


The Central Limit Theorem in it's simplest form states that whenever the mean is taken for random samples from a population then the distribution of those means tend to follow a normal distribution<br> 

Furthermore the mean if these new computed means of these sample tends to equal to the mean of the population.

Therefore the mean of the sample means is given as:- 

$$
\mu_\bar{x} = \mu 
$$


The variance of the sample means is given as 

$$
\sigma_\bar{x} =  \frac{\sigma}{\sqrt{n}}
$$


Here :- 


$\mu_\bar{x}$  is the mean of the sample means.<br>
$\mu$ is the parameter mean i.e. population mean.<br>
$\sigma_\bar{x}$ is the variance of the sample means.<br>
$n$ is the number of observations per sapmle selected.<br>
$\sigma $ is the parameter variance i.e. population variance.<br>




Lets get a clearer understanding of this in python


```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
import seaborn as sns
import random 
```

Lets generate a population with $\mu$  = 5.0 and $\sigma $ = 1.0.<br>
There we be a total of 100000 observations.<br>
Let's also plot the distribution for the same.<br>

```python
mu, sigma = 5.0, 1.0 # mean and standard deviation
s = np.random.normal(mu, sigma, 100000)
```

```python
sns.distplot(s)
plt.show()
```

Lets take a total of 20 samples from the population with 10000 observations.<br>
We will compute the means and variance and plot the distribution.<br>

```python
sample_means_1=[]

for i in range(20):
        sapmle_list=random.sample(np.array(s).tolist(), 10000)
        sample_means_1.append(np.mean(sapmle_list))
    
```

```python
sns.distplot(sample_means_1)
plt.show()
```

Lets compute the mean and variance for above list from the formules given above as well as through computation.

```python
print('Mean :- '+str(np.mean(sample_means_1)))
print('Population Mean :- '+str(5.0))
print('Variance :- '+str(np.std(sample_means_1)))
print('Variance By formulae:- '+str(1/np.sqrt(10000)))
```

Thus we see the results are matching.<br>
Similarly lets go throught some meore variations.<br>

```python
sample_means_2=[]

for i in range(20000):
        sapmle_list=random.sample(np.array(s).tolist(), 5)
        sample_means_2.append(np.mean(sapmle_list))
    
```
